```python
import geocoder

#map
parana = folium.Map(
    location=[-25.4440699,-50.1277363],
    zoom_start=8
)

#Pegando a localizacao do endereco
address = geocoder.osm('cidade de Curitiba')

#Endereço latitude e longitude 
address_latlon = [address.lat, address.lng]

#add marker to map
folium.Marker(address_latlon, popup='Curitiba', tooltip='click').add_to(parana)

#display map
parana
```
