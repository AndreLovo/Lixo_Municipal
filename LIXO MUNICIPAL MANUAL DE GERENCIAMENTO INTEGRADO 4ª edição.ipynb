```python
arquivo_texto=arquivo_texto['Ano;Faturamento;Capacidade_produÃ§Ã£o;Investimento;ExportaÃ§Ãµes'].str.split(';', expand=True)
display(arquivo_texto)
```

```python
#Renomeando as colunas
arquivo_texto.rename(columns={0: 'ano'}, inplace=True)
arquivo_texto.rename(columns={1: 'faturamento'}, inplace=True)
arquivo_texto.rename(columns={2: 'capacidade_producao'}, inplace=True)
arquivo_texto.rename(columns={3: 'investimento'}, inplace=True)
arquivo_texto.rename(columns={4: 'exportacao'}, inplace=True)
display(arquivo_texto)
```

```python
# Verificando as informações da tabela 
arquivo_texto.info()
```

```python
#Alterando o formato das colunas
#object para int

arquivo_texto['ano'] = arquivo_texto['ano'].astype(int)
arquivo_texto['faturamento'] = arquivo_texto['faturamento'].astype(int)
arquivo_texto['capacidade_producao'] = arquivo_texto['capacidade_producao'].astype(int)
arquivo_texto['investimento'] = arquivo_texto['investimento'].astype(int)
arquivo_texto['exportacao'] = arquivo_texto['exportacao'].astype(int)
```

```python
# Aumentando o tamanho dos gráficos
plt.figure(figsize=(16,10))
#Plotando o gráfico de barras
plt.bar(arquivo_texto.ano, arquivo_texto.faturamento)
plt.xticks (rotation= 0, fontsize=9)
plt.title("Quadro 1: ano x faturamento")
plt.show()
```

```python
# Imprimindo o gráfico scatter - relplot
plt.figure(figsize=(20,20))
sns.relplot(x="ano", y="faturamento", data=arquivo_texto, size="faturamento", sizes=(12,250))
plt.xticks(color='blue',rotation=0,size=9)
plt.show()
```

```python
plt.figure(figsize=(6,18))

ano = arquivo_texto ['ano'].tolist()
faturamento = arquivo_texto ['faturamento'].tolist()
investimento = arquivo_texto ['investimento'].tolist()
exportacao = arquivo_texto ['exportacao'].tolist()

plt.plot(ano, faturamento, label = 'faturamento', color='g', marker='o', markerfacecolor='k', linestyle='--', linewidth=3)
plt.plot(ano, investimento, label = 'investimento', color='r', marker='o', markerfacecolor='k',linestyle='--',linewidth=3)
plt.plot(ano, exportacao, label = 'exportacao', color='b', marker='o', markerfacecolor='k',linestyle='--',linewidth=3)  
    
plt.xlabel('ano')
plt.ylabel('valores')
plt.legend(loc='upper right')
plt.title(' ')
plt.xticks(ano)
plt.yticks([9, 17, 19, 27, 29, 46, 72, 73, 83, 91, 95, 108, 896, 1119, 1078, 1081, 759, 789])
plt.show()
```

```python
# Imprimir gráfico de Pizza para verificação de Porcentagem
ano=[2003,2004,2005,2006,2007,2008]
faturamento=[896,1119,1078,1081,759,789]

explode = (0.0,0.3,0,0.1,0,0)

# Criando a figura para a plotagem
plt.subplots(figsize=(7,7))

# Fazendo um gráfico de Pizza para melhor visualizacao
plt.pie(x= faturamento, labels= ano, explode=explode, autopct='%1.1f%%', startangle=90)

# Mostrando o grafico
plt.show()
```
